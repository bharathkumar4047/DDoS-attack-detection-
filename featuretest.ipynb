```python
import pandas as pd
```

```python
df = pd.read_csv("testingdata.csv")
```

```python
df.columns = df.columns.str.strip()

# Remove rows where Label is 1.0, 2.0, 3.0, or 5.0
df = df[~df["Label"].isin([1.0, 2.0, 3.0, 5.0])]

# Save to new CSV (optional)
df.to_csv("test_cleaned_file.csv", index=False)
```

```python


# Clean column names (optional)
df.columns = df.columns.str.strip()

# Convert label 4.0 → 1.0
df["Label"] = df["Label"].replace(4.0, 1.0)

# Save the updated file
df.to_csv("test_updated_labels.csv", index=False)
```
