```python
import pandas as pd
import numpy as np
from tensorflow.keras.models import load_model

# ------------------------------
# LOAD THE TRAINED MODEL
# ------------------------------
MODEL_PATH = "ddos_model.h5"
model = load_model(MODEL_PATH)

# ------------------------------
# LOAD CSV FILE
# ------------------------------
CSV_PATH = "test_updated_labels.csv"
df = pd.read_csv(CSV_PATH)

# ------------------------------
# REMOVE LABEL COLUMN
# ------------------------------
# Replace "Label" with your actual label column name
label_column = "Label"  

if label_column in df.columns:
    df = df.drop(columns=[label_column])

# ------------------------------
# TAKE ONLY FIRST ROW
# ------------------------------
row = df.iloc[600:601]
print("\n--- FIRST ROW (WITHOUT LABEL) ---")
print(row)

# ------------------------------
# SELECT NUMERIC FEATURES ONLY
# ------------------------------
features = row.select_dtypes(include=['int64', 'float64'])
X = features.to_numpy()

# If your model used StandardScaler:
# from joblib import load
# scaler = load("scaler.pkl")
# X = scaler.transform(X)

# ------------------------------
# PREDICT
# ------------------------------
prediction = model.predict(X)[0][0]  # probability
label = "ATTACK" if prediction > 0.5 else "NORMAL"

print("\n--- RESULT ---")
print("Prediction:", label)
print("Probability:", prediction)
```
