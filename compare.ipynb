```python
import matplotlib.pyplot as plt

# =========================
# Replace with your values
# =========================
xgb_accuracy = 0.98
dl_accuracy = 0.99

models = ['XGBoost', 'Deep Learning']
accuracy_values = [xgb_accuracy, dl_accuracy]

plt.figure(figsize=(4,4))
plt.bar(models, accuracy_values)
plt.ylabel('Accuracy')
plt.ylim(0, 2)
plt.title('Accuracy Comparison')
plt.grid(axis='y', linestyle='--', alpha=0.4)

# Show values above bars
for i, acc in enumerate(accuracy_values):
    plt.text(i, acc + 0.01, f"{acc:.2f}", ha='center', fontsize=11)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# ==============================
# Replace these with your values
# ==============================
xgb_accuracy = 0.9856
dl_accuracy = 0.9997

xgb_precision = 0.9869
dl_precision = 1.0

xgb_recall = 0.9905
dl_recall = 0.9995

xgb_f1 = 0.9887
dl_f1 = 0.9998

# ==============================
# Create the graph
# ==============================

metrics = ["Accuracy", "Precision", "Recall", "F1 Score"]
xgb_scores = [xgb_accuracy, xgb_precision, xgb_recall, xgb_f1]
dl_scores = [dl_accuracy, dl_precision, dl_recall, dl_f1]

x = np.arange(len(metrics))
width = 0.35

plt.figure(figsize=(8,5))
plt.bar(x - width/2, xgb_scores, width, label='XGBoost')
plt.bar(x + width/2, dl_scores, width, label='Deep Learning')

plt.xticks(x, metrics)
plt.ylabel("Score")
plt.ylim(0, 1.5)
plt.title("XGBoost vs Deep Learning: Performance Comparison")
plt.legend()

# Print score labels above bars
for i, v in enumerate(xgb_scores):
    plt.text(i - width/2, v + 0.01, f"{v:.2f}", ha='center', fontsize=10)

for i, v in enumerate(dl_scores):
    plt.text(i + width/2, v + 0.01, f"{v:.2f}", ha='center', fontsize=10)

plt.show()
```
